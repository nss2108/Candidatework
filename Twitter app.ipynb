```python
import re
from textblob import TextBlob
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
def get_tweet_sentiment(tweet):
    analysis=TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return "positive"
    if analysis.sentiment.polarity == 0:
        return "neutral"
    if analysis.sentiment.polarity < 0:
        return "negative"

```

```python
def sent_analysis_df(df):
    clean_tweets=[]
    for tweet in df.tweetText:
        clean_tweets.append(clean_tweet(tweet))
    sentiments=[]
    for tweet in clean_tweets:
        sent=get_tweet_sentiment(tweet)
        sentiments.append(sent)
    df["sentiments"]=sentiments
    ptweets = df[df['sentiments'] == 'positive']
    # percentage of positive tweets
    print("Positive tweets percentage: {} %".format(100*len(ptweets)/len(df)))
    # picking negative tweets from tweets
    ntweets = df[df['sentiments'] == 'negative']
    # percentage of negative tweets
    print("Negative tweets percentage: {} %".format(100*len(ntweets)/len(df)))
    # percentage of neutral tweets
    neuttweets=df[df["sentiments"] == "neutral"]
    print("Neutral tweets percentage: {} % \
        ".format(100*len(neuttweets)/len(df)))
    return df
sentdf=sent_analysis_df(df)
    
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from gensim.corpora.dictionary import Dictionary
from nltk.tokenize import word_tokenize
```

```python
import re
tweets=df.tweetText.tolist()
from nltk.tokenize import TweetTokenizer
tt=TweetTokenizer(strip_handles=True, preserve_case=False)
tokentweets=[]
for i in tweets:   
    tokentweets.append(tt.tokenize(i))
tokentweets=list(map( " ".join, tokentweets))
tokentweets=[re.sub(r"http\S+", "", x) for x in tokentweets] #deletes urls
tokentweets=[re.sub(r"rt : ", "", x) for x in tokentweets] #deletes rts
tokentweets=[re.sub(r"[^a-zA-Z0-9'_*.\s]", "", x) for x in tokentweets] #eliminates non-standard characters
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
tokentweets=[i for i in tokentweets if i not in stop]
```

```python
tk=list(map(word_tokenize, tokentweets))
dictionary=Dictionary(tk)
corpus = [dictionary.doc2bow(text) for text in tk]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def make_nb(Xvals, yvals):
    vectorizer = CountVectorizer(min_df=0)
    vectorizer.fit(tokentweets)
    x = vectorizer.transform(Xvals)
    y=(yvals["sentiments"]=="positive").values.astype(np.int)
    return x, y
X, y=make_nb(tokentweets, sentdf)
```

```python
from sklearn.model_selection import KFold
kfold=KFold(n_splits=5, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
parameters={"alpha": [.001,.01,.1, 1, 5, 10]}
nb=MultinomialNB()
model=GridSearchCV(nb, parameters, cv=kfold, scoring="accuracy")
model.fit(X_train, y_train)
bestalpha=model.best_params_["alpha"]
```

```python
clf=MultinomialNB(alpha=bestalpha)
clf.fit(X_train, y_train)
training_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)

print("Accuracy on training data: {:2f}".format(training_accuracy))
print("Accuracy on test data:     {:2f}".format(test_accuracy))
```

```python
words = np.array(vectorizer.get_feature_names())

x = np.eye(X_test.shape[1])
probs = clf.predict_log_proba(x)[:, 0]
ind = np.argsort(probs)

good_words = words[ind[:20]]
bad_words = words[ind[-20:]]

good_prob = probs[ind[:20]]
bad_prob = probs[ind[-20:]]

print("Positive Sentiments     P(positive | word)")
for w, p in zip(good_words, good_prob):
    print("{:>20}".format(w), "{:.2f}".format(1 - np.exp(p)))
    
print("Negative/Neutral Sentiments  P(negative/neutral | word)")
for w, p in zip(bad_words, bad_prob):
    print("{:>20}".format(w), "{:.2f}".format(1 - np.exp(p)))
```
